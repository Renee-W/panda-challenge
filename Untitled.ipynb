```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set x axis to numerical value for month
x_axis = np.arange(1,13,1)
x_axis
```

```python
# Avearge weather temp
points_F = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
```

```python
# Convert to Celsius C = (F-32) * 0.56
points_C = [(x-32) * 0.56 for x in points_F]
points_C
```

```python
plt.plot(x_axis, points_F)
plt.plot(x_axis, points_C)
plt.show()
```

```python
f_handle, = plt.plot(x_axis, points_F, marker ='+', color="blue",  label = "Temp in F")
c_handle, = plt.plot(x_axis, points_C, marker = 's', color = 'red', label = "Temp in C")
plt.title("Temp in NJ")
plt.xlabel("Months")
plt.ylabel("Temp")
plt.legend(loc='lower right')
```
