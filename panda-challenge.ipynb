```python
#import pandas
import pandas as pd
```

```python
#path to csv file
school = "PyCitySchools/Resources/schools_complete.csv"
student = "PyCitySchools/Resources/students_complete.csv"

#read files
school_df = pd.read_csv(school)
student_df = pd.read_csv(student)
```

```python
school_df['budgetsd']=(school_df['budget']/school_df['size'])
```

```python
#merge tables
studentschool_df=pd.merge(student_df, school_df, on='school_name', how='left')
```

```python
#calculate passing scores
convert_passing_df = studentschool_df[['student_name', 'grade', 'school_name', 'reading_score', 'math_score']]
passing_df = convert_passing_df.rename(columns={'math_score':'math_pass', 
                                                       'reading_score':'reading_pass'})
```

```python
passing_df.loc[passing_df['math_pass']<70, 'math_pass']=0
passing_df.loc[passing_df['reading_pass']<70, 'reading_pass']=0
passing_df.loc[passing_df['math_pass']>69, 'math_pass']=1
passing_df.loc[passing_df['reading_pass']>69, 'reading_pass']=1
mathpass=(passing_df['math_pass']).sum()
readpass=(passing_df['reading_pass']).sum()
```

```python
#calculate math & reading pass
passing_df["Passing"] = passing_df.sum(axis=1)
```

```python
#calculate total
passing_df.loc[passing_df['Passing']<2, 'Passing']=0
passing_df.loc[passing_df['Passing']>1, 'Passing']=1
OverallPassing=(passing_df['Passing']).sum()
```

```python
### District Summary
#Total Schools
SchoolName = len (studentschool_df['school_name'].drop_duplicates())
#Total Students
TotalStudent = len(studentschool_df.index)
#Total Budget
TotalBudget = round(studentschool_df["budget"]).sum()
#Average Math Score
Math = round((studentschool_df["math_score"]).mean(),2)
#Average Reading Score
Reading = round((studentschool_df["reading_score"]).mean(),2)
#% Passing Math (The percentage of students that passed math)
mathpercent = round(((mathpass/TotalStudent)*100),2)
#% Passing Reading (The percentage of students that passed reading.)
readpercent = round(((readpass/TotalStudent)*100),2)
#% Overall Passing (The percentage of students that passed math **and** reading.)
passingpercent = round(((OverallPassing/TotalStudent)*100),2)

#Summary
summary_df=pd.DataFrame({"Total Schools": [SchoolName],
                         "Total Students":[TotalStudent], 
                        "Total Budget":[TotalBudget],
                        "Average Math Score":[Math],
                        "Average Reading Score": [Reading],
                        "% Passing Math":[mathpercent],
                        "% Passing Reading":[readpercent],
                        '% Overall Passing':[passingpercent]})

summary_df["Total Budget"]=summary_df["Total Budget"].astype(float).map('${:,.2f}'.format)
summary_df
```

```python
#stats for school

schoolstat_df = studentschool_df.groupby(["school_name"])
school1_df=round(schoolstat_df.mean(),2)

schoolpass_df=passing_df.groupby(['school_name'])
school2_df=schoolpass_df.sum()

school3_df=pd.merge(school_df, school2_df, on="school_name", how='inner')
school3_df['reading_percent'] = round(((school3_df['reading_pass']/school3_df['size'])*100),2)
school3_df['math_percent'] = round(((school3_df['math_pass']/school3_df['size'])*100),2)
school3_df['total_percent'] = round(((school3_df['Passing']/school3_df['size'])*100),2)

schoolsum1_df=pd.merge(school1_df, school3_df, on="school_name", how='inner')
```

```python
### School Summary
# Create an overview table that summarizes key metrics about each school, including:
  # School Type
  # Total Students
  # Total School Budget
  # Per Student Budget
  # Average Math Score
  # Average Reading Score
  # % Passing Math (The percentage of students that passed math.)
  # % Passing Reading (The percentage of students that passed reading.)
  # % Overall Passing (The percentage of students that passed math **and** reading.
schoolsummary_df = schoolsum1_df[['school_name', 'type', 'size_y', 'budget_y', 'budgetsd_y',
                                  'math_score', 'reading_score', 'math_percent', 'reading_percent', 'total_percent']]
schoolsummary_df =schoolsummary_df.rename(columns = {'school_name':'School', 'type':'School Type', 'size_y':'Total Students',
                                  'budget_y':'Total School Budget', 'budgetsd_y':'Per Student Budget',
                                  'math_score':'Average Math Score', 'reading_score':'Average Reading Score',
                                  'math_percent':'% Passing Math', 'reading_percent':'% Passing Reading', 
                                  'total_percent':'% Overall Passing'})
schoolsummary_df = schoolsummary_df.set_index("School")
schoolsummary_df["Total School Budget"]=schoolsummary_df["Total School Budget"].astype(float).map('${:,.2f}'.format)
schoolsummary_df["Per Student Budget"]=schoolsummary_df["Per Student Budget"].astype(float).map('${:,.2f}'.format)

schoolsummary_df
```

```python
### Top Performing Schools (By % Overall Passing)
schoolsummary_df=schoolsummary_df.sort_values("% Overall Passing", ascending=False)
schoolsummary_df.head(5)
# Create a table that highlights the top 5 performing schools based on % Overall Passing. Include:
#  * School Name
#  * School Type
#  * Total Students
#  * Total School Budget
#  * Per Student Budget
#  * Average Math Score
#  * Average Reading Score
#  * % Passing Math (The percentage of students that passed math.)
#  * % Passing Reading (The percentage of students that passed reading.)
#  * % Overall Passing (The percentage of students that passed math **and** reading.)
```

```python
### Math Scores by Grade\*\*
# Create a table that lists the average Math Score for students of each grade level (9th, 10th, 11th, 12th) at each school.
#studentschool_df.head()
math_df = studentschool_df[['student_name', 'grade', 'school_name', 'math_score']]

math9_df = math_df.loc[math_df["grade"] == "9th"]
math9stat1_df = math9_df.groupby(["school_name"])
math9stat_df=round(math9stat1_df.mean(),2)
math9stat_df = math9stat_df.rename(columns = {'math_score':'9th'})

math10_df = math_df.loc[math_df["grade"] == "10th"]
math10stat1_df = math10_df.groupby(["school_name"])
math10stat_df=round(math10stat1_df.mean(),2)
math10stat_df = math10stat_df.rename(columns = {'math_score':'10th'})

math11_df = math_df.loc[math_df["grade"] == "11th"]
math11stat1_df = math11_df.groupby(["school_name"])
math11stat_df=round(math11stat1_df.mean(),2)
math11stat_df = math11stat_df.rename(columns = {'math_score':'11th'})

math12_df = math_df.loc[math_df["grade"] == "12th"]
math12stat1_df = math12_df.groupby(["school_name"])
math12stat_df= round(math12stat1_df.mean(),2)
math12stat_df = math12stat_df.rename(columns = {'math_score':'12th'})

mathgd1_df=pd.merge(math9stat_df, math10stat_df, on="school_name", how='inner')
mathgd2_df=pd.merge(math11stat_df, math12stat_df, on="school_name", how='inner')
mathgrade_df = pd.merge(mathgd1_df, mathgd2_df, on="school_name", how='inner')

mathgrade_df
```

```python
### Reading Scores by Grade
reading_df = studentschool_df[['student_name', 'grade', 'school_name', 'reading_score']]

reading9_df = reading_df.loc[reading_df["grade"] == "9th"]
reading9stat1_df = reading9_df.groupby(["school_name"])
reading9stat_df= round(reading9stat1_df.mean(),2)
reading9stat_df = reading9stat_df.rename(columns = {'reading_score':'9th'})

reading10_df = reading_df.loc[reading_df["grade"] == "10th"]
reading10stat1_df = reading10_df.groupby(["school_name"])
reading10stat_df= round(reading10stat1_df.mean(),2)
reading10stat_df = reading10stat_df.rename(columns = {'reading_score':'10th'})

reading11_df = reading_df.loc[reading_df["grade"] == "11th"]
reading11stat1_df = reading11_df.groupby(["school_name"])
reading11stat_df= round(reading11stat1_df.mean(),2)
reading11stat_df = reading11stat_df.rename(columns = {'reading_score':'11th'})

reading12_df = reading_df.loc[reading_df["grade"] == "12th"]
reading12stat1_df = reading12_df.groupby(["school_name"])
reading12stat_df= round(reading12stat1_df.mean(),2)
reading12stat_df = reading12stat_df.rename(columns = {'reading_score':'12th'})

readinggd1_df=pd.merge(reading9stat_df, reading10stat_df, on="school_name", how='inner')
readinggd2_df=pd.merge(reading11stat_df, reading12stat_df, on="school_name", how='inner')
readinggrade_df = pd.merge(readinggd1_df, readinggd2_df, on="school_name", how='inner')

readinggrade_df
```

```python
### Scores by School Spending

# Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
#  * Average Math Score
#  * Average Reading Score
#  * % Passing Math (The percentage of students that passed math.)
#  * % Passing Reading (The percentage of students that passed reading.)
#  * % Overall Passing (The percentage of students that passed math **and** reading.)
schoolsummary2_df = schoolsum1_df[['school_name', 'type', 'size_y', 'budget_y', 'budgetsd_y',
                                  'math_score', 'reading_score', 'math_percent', 'reading_percent', 'total_percent']]
schoolsummary2_df =schoolsummary2_df.rename(columns = {'school_name':'School', 'type':'School Type', 'size_y':'Total Students',
                                  'budget_y':'Total School Budget', 'budgetsd_y':'Per Student Budget',
                                  'math_score':'Average Math Score', 'reading_score':'Average Reading Score',
                                  'math_percent':'% Passing Math', 'reading_percent':'% Passing Reading', 
                                  'total_percent':'% Overall Passing'})

spendingsum_df=schoolsummary2_df[['Per Student Budget', 'Average Math Score', 'Average Reading Score', '% Passing Math', 
                              '% Passing Reading', '% Overall Passing']]
# Create bins in which to place values based upon school spending
bins = [0, 585, 630, 645, 680]

# Create labels for these bins
group_labels = ["<585", "585 to 630", "630 to 645", "645 to 680"]
```

```python
#data into bins
spendingsum_df['Per Student Summary']=pd.cut(spendingsum_df["Per Student Budget"], bins, labels=group_labels, include_lowest=True)
```

```python
# data into new colulmn
#spendingsum_df['Per Student Summary']= pd.cut(spendingsum_df["Per Student Budget"], bins, labels=group_labels)

spendsumgp_df= spendingsum_df[['Per Student Summary', 'Average Math Score', 'Average Reading Score', 
                               '% Passing Math', '% Passing Reading', '% Overall Passing']]
spendgp1_df=spendsumgp_df.groupby(['Per Student Summary'])
spendgp2_df=round(spendgp1_df.mean(),2)
spendgp2_df
```
